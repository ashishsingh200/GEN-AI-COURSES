```python
try:
    #Taking 2 inputs from user
    numerator=int(input("Enter the Numerator value:"))
    denominator=int(input("Enter the denominator:"))    
    result=numerator/denominator #finding the division from the given input
except ZeroDivisionError as e1: #checking ZeroDivisionError
    print(e1)
except ValueError as e2: #checking ValueError
    print(e2)
else:
    print(result) #Printing the results
finally:
    print('Operation Complete') #finally
```

```python
prices=[120,350,'abc',500,-200,800] #list containing prices of products
sum=0
for price in prices:
    try:
        if not isinstance(price,(int,float)): #checking the instance 
            raise TypeError("Value is not a number")
        if price < 0: 
            raise ValueError("Negative price not allowed")
        sum += price
    except TypeError as e1:
        print(f"Skipping Invalid Type: {price}")
    except ValueError as e2:
        print(f'Skipping Invalid Value: {e2}')
print(sum)
```

```python
def check_age(age): #defining a function to check wether the given age is valid or not
    if age < 1 and age >120:
        raise ValueError("Age must be between 1 and 120")
    else:
        print('Age is Valid!')

try:
    age=int(input("Enter the age:"))
    check_age(age)
except Exception as e:
    print(e)
```

```python
cart=[]
bill=0
while True:
    item=input("Enter the Prices or enter q to exit.")
    if item=='q': break
    else:
        try:
            f_item=float(item)
            if f_item<0: 
                raise ValueError("the given price is negative")
        except ValueError as e1:
            print(e1)
        else:
            cart.append(f_item)
for item in cart:
    bill +=item
print(f"Total Items: {len(cart)}")
print(f'Total Bill: {bill}')
```
