## Exception Handling <---> Error Handling Process

#### Eception Handling is a way which if wwe are getting the errors in programming or if the  user enterd input causes an error, then instead of throwing an Error we can raise an exception which can let the user know that the given input is not correct or the given input can throww an error.

##### Exception are events that disrupt the normal flow of a program. They occur when an error is encountered during program execution.

## Different Type of Exception

- ZeroDivisionError --> Dividing by zero
- FileNotFoundError --> File not found
- ValueError --> Invalid Value
- TypeError --> Invalid Type

```python
try:
    num=int(input("Enter the Denominator: "))
    result=1/num
    print(result)
except ZeroDivisionError:
    print("Number Can't be divided by Zero")
except ValueError:
    print("Enter the correct va lue in integer...")
```

### try/except/else

```python
num=int(input('Enter the number'))
try:
    result=1/num
except Exception as e:
    print(e)
else:
    print(result)
```

### try/except/else/finally

```python
try:
    num = int(input("Enter a number: "))
    result = 100 / num
except ValueError:
    print("Please enter a valid number!")
except ZeroDivisionError:
    print("Cannot divide by zero!")
else:
    print(f"Result: {result}")      # only if no error
finally:
    print("Program complete.")      # always runs
```
