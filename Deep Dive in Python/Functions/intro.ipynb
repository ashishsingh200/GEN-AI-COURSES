# Functions

- A function is a block of code that performs a specific task.
- Functions helps in organizing code, Reusing code and improving readability

```python
def add(a,b):
    return a+b
```

```python
a=int(input("Enter 1st number: "))
b=int(input("Enter 2nd number: "))
res=add(a,b)
print(f'{a} + {b} = {res}')
```

```python
def even_odd(number):
    if not number.isdigit():
        print("INVALID INPUT TYPE!")
        return
    else:
        if number %2 ==0:
            return "Even Number"
        else: return "Odd Number"

num=input("Enter the number to check wether it is even or odd: ")
print(f'The number {num} is {even_odd(num)}')
```

```python
print(even_odd('s'))
```

```python
def even_odd_instance(num):
    if isinstance(num,int):
        if num %2 ==0: print(f"{num} is an Even Number")
        else: print(f'{num} is Odd Number')
    elif isinstance(num,str):
        print("Invalid data type...")
    else:
        print(f'{num} is an invalid value for this function.')
```

```python
even_odd_instance('p')
```

```python
even_odd_instance(True)
```
