```python
# To create a function that converts temprature fro celcius to fahrenheit and vice-versa


def convert(temp,unit):
    if isinstance(unit,str) and len(unit)==1:
        if unit.lower()=='c':
            result=(temp-32) *(5/9)
            return print(f'{temp} Fahrenheit equals to {result} Celcius')
        
        elif unit.lower()=='f':
            result=(temp * (9/5)) + 32
            return print(f'{temp} Celcius equals to {result} Fahrenheit')
        else: print('Enter the correct unit either F or C')

```

```python
convert(1,'f')
```

```python
convert(1,'c')
```

```python
# define a function that would behave as. a password strebgth checker

'''
    - for a strong password -->
        1. length should be greater than 8 character.
        2. password must consist of atleast of 1 lower case letter(a,b,c,d,...)
        3. password must consist of atleast of 1 upper case letter(A,B,C,...)
        4. password must consist of atleast of 1 digit letter(1,2,3,4,...)
        5. password must consist of atleast of 1 special character(!,<,@,#,...)
    
    - Weak Password -->
        if it doesn't satisfy the above condition.
'''
import string
def check_pass(password):
    if len(password) < 8 : 
        print(f' The password You enterd {password} is weak.')
    elif not any(char.islower() for char in password):
        print(f' The password You enterd {password} is weak.')
    elif not any(char.isupper() for char in password):
        print(f' The password You enterd {password} is weak.')
    elif not any(char.isdigit() for char in password):
        print(f' The password You enterd {password} is weak.')
    elif not any(char in string.punctuation for char in password):
        print(f' The password You enterd {password} is weak.')
    else: 
        print(f' The password You enterd {password} is Strong.')
    

    
```

```python
check_pass('Singh@1234')
```

```python
check_pass('H0000')
```

```python
check_pass('Ashish222')
```

```python
# Check wether the given string is Pailindrome or not
''' 
    Pailindrome words are like 
    - aba and aba
    -adfda and adfda

    meaning words who sound and spell same if we start reading from backwards
'''
```

```python
def isPailindrome(words):
    if not isinstance(words, str):
        print('We need words not any other thing the input type is str....')

    else:
        result=words[::-1]
        if result==words: print(f'{words} is a pailindrome')
        else: print(f'{words} is not pailindrome')
```

```python
isPailindrome('aba')
```

```python
def ispalindrome(words):
    
    n=len(words)
    for i in range(len(words)//2):
        if  words[i] != words[n-1]: 
            return False
        n-=1
    return True
```

```python
ispalindrome('abcba')
```

```python
ispalindrome('jjj')
```

```python
ispalindrome('kkjs j')
```
