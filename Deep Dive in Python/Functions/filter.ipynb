# filter()

```python
numbers=[1,2,3,4,5,6,7,8]
```

```python

def is_even(n):
    return n%2 == 0


result=list(filter(is_even,numbers))

result
```

```python
result=list(filter(lambda n:n%2==0,numbers))
print(result)
```

```python
#getting the items >5
result=list(filter(lambda x: x>5 and x%2==0,numbers))
result
```

```python
#getting the items >5
result=list(map(lambda x: x*x,numbers))
result
```

```python
# iterating over the dictionary items


emp=[
    {'name':'Amit','age':24,'qualification':'Bachelor'},
    {'name':'Ashish','age':21,'qualification':'Bachelor'},
    {'name':'Aman','age':20,'qualification':'Bachelor'},
    {'name':'Abhishek','age':24,'qualification':'BBA'},
    {'name':'Aditya','age':29,'qualification':'Master'}
]
def is_eligible(emp):
    if emp['age']>21:
        return True
result=list(filter(is_eligible,emp))
result
```
