```python
def power(num,exp):
    if not (isinstance(num,int) and isinstance(exp,int)):
       print('Invalid Format it should be power(int,int)')
       return
    else:
        if exp<0:
            print("Exponent must be greater than 0.")
            return
        elif exp:=0: 
            return 1
        
        return num * power(num,exp-1)
```

```python
power(2,3)
```

```python
power(2,-3)
```

```python
power(-2,3)
```

```python
power(2,0)
```

```python
def display(**student_data):
    for key, values in student_data.items():
        print(f'{key}: {values}')
```

```python
display(name='Ashish',age='25',height='6 feet')
```

```python
def variable(*positional,**keyword):
    students=keyword.get('students',[])
    for i,student in zip(positional,students):
        print(f'Roll of student is: {i}')
        for key , values in student.items():
            print(f'{key}: {values}')
        print()
```

```python
variable(1,2,3,4,5,
        students=[
            {'name':'Ashish','age':24 }, 
            {'name':'prince','age':90},
            {'name':'Badal','age':45},
            {'name':'Ramesh','age':32},
            {'name':'Riya', 'age':94}
        ])
```
