# Lambda Functions

```python
# lambda function to add two numbers and comparing with regular expressions


# regular function
def add(a,b):
    return a+b

# lambda function

add2=lambda a,b:a+b
```

```python
add2(2,4)
```

```python
add(3,5)
```

```python
power=lambda a,b:a**b
```

```python
power(2,3)
```
