# File Handling

## open() --> to open a file
#### parameters --> file name
#### second parameter --> mode
    - r -->reading the content from my text file
    - w --> writing the content in the file
    - r+ --> read the content first and then writing the context into the file
    - w+ --> writing the file first and then reading from it

##### writing the content into the file

    - write() --> it overwrites the content into the text file
    - append() --> it adds the data from the end of the file

```python
#reading the content from my text file

f=open('example.txt','r')
content=f.read()
print(content)
f.close()
```

```python
# writing in the file

#f=open('example.txt','w')
#f.write('yoo')
```

```python
# appending the content into the file
f=open('example.txt','a')
f.write('174 cm')
f.close()
```

## with --> it will be used for first opening the file as a object there is no requirement for closing the file as well

```python
with open('example.txt','r') as f:
    content_data=f.read()
    print(content_data)
```

```python
with open('example.txt','r+') as f:
    content_data=f.read()
    print(content_data)

    f.write('\nAshish Kumar Singh')
    print(content_data)
```

```python
with open('example.txt','w') as file:
    file.writelines(['As','\nhi'])
```

```python
products = ['pen\n', 'copy\n', 'macbook\n']
with open('products.txt', 'w') as file:
    file.writelines(products)
```

# Reading from a file

```python
with open('products.txt','r') as file:
    content= file.read()
    print(content)
```

```python
with open('products.txt','r') as file:
    line=file.readline()
    print(line)
```

```python
with open('products.txt','r') as file:
    lines=file.readlines()
    print(lines)
```

```python
with open('products.txt','r') as file:
    for line in file:
        print(line.strip())
```

# Appending  to a File

```python
with open('products.txt','a') as file:
    file.write('bottle\n')
```

# Checking if a file exist or not

```python
import os
if os.path.exists('products.txt'):
    with open('products.txt','r') as file: 
        print(file.read())
else: print('File Not Found!')
```

# File pointer 

    - seek()

    - tell()

```python
with open('products.txt','r') as file:
    print(file.tell())

    print(file.read(3))

    print(file.tell())

    print(file.seek(4))

    print(file.tell())

    print(file.read())
```

# deleting a file

```python
import os
if os.path.exists('example.txt'):
    os.remove('example.txt')
    print('File Deleted')

else:
    print('file not esist!')
```
