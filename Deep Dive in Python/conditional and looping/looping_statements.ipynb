## Loop control statements

```python
for i in range(51):
    if i==18: break
    else: print(i)
```

```python
for i in range(5):
    if i==3: continue
    print(i)
```

```python
for i in range(5):
    if i==3: 
        print("I don't like number 3")
        pass
    else: 
        print(i)
```

```python
#print prime from 0 to 100

for num in range(2,101):
    for i in range(2,num):
        if num % i==0: break
    else: print(num)
```
