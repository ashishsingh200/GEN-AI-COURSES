# Conditional Statements

- Conditional statements works like Real Life scenario where if a certain condition is meeting then a statement of code or a block of code gets executed

```python
# Eligibility for driving on meeting a certain condition

age=int(input("Enter Your age: "))

if age >=18: print('You are Eligible For Driving')
else: print('You are not Eligible for Driving')
```

```python
# figure out the tallest among 3 friends

amit=int(input("Enter your height: "))
ashish=int(input("Enter your height: "))
priya=int(input("Enter your height: "))

if amit>ashish and amit>priya:
    print("Amit is tallest!")
elif ashish>amit and ashish>priya:
    print("Ashish is tallest!")
else:
    print('Priya is tallest!')
```

```python
## wether if the current year is a leap year or not

year=int(input("Enter the Year: "))

if year % 4==0:
    if year % 100 == 0:
        if year % 400==0:
            print(f'{year} is a leap year')
        else:
            print(f'{year} is not a leap year')
    else:
        print(f'{year} is a leap year')
else: 
    print(f'{year} is not a leap year')
```
