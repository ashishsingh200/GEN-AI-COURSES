Type Casting (also called Type Conversion) in Python means changing a value from one data type to another.
Since Python is dynamically typed, it often does this automatically, but sometimes you need to do it manually.

There are two main types:

Implicit Type Conversion (automatic, done by Python)

Explicit Type Conversion (manual, done by you using functions)

```python
a=6
b=7
print(a+b)
print(type(a))
```

```python
a=str(a)

print(a+b)
type(a)
```
