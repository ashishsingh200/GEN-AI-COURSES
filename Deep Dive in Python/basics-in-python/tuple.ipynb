There are two methods in tuples

    - index()

    -count()

```python
a=(1,2,3,2,1,2,3,4,2,1,3,4,2,1,1)
a.index(3) #returns the element present in that index
```

```python
a.count(1) #return the no. of times the element 1 present in tuple a
```

- Iterating over a Tuple inside a Tuple

```python
tuple_list
```

```python
for i in tuple_list:
    for j in i:
        print(j,sep="")
    print()
```
