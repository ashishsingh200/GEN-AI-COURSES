# Polymorphism

## Polymorphism --> means many forms in greek. 
## In OOPs, Polymorphism allows objects of different classes as if they are the object of the same class and it enables reusability and makes programs more flexible.

### Types of Polymorphism
1. Method Overriding (Runtime "Polymorphism)
2. Method Overloading (Compile Time Polymorphism)
3. Operator Overloading

#### Method Overriding --> A child class provides a specific implementation of a method that has already been defined into the parent class. This allows Dynamic Selection at Runtime

```python
class GrandParent: #Parent class
    def speak(self):
        print("Hello my name is Ashish")

class Parent(GrandParent): #child class
    def speak(self):
        print("Hello my name is Ashish jr.")
class Child(GrandParent): #grand child class
    def speak(self):
        print("Love you Old Man")
```

```python
obj1=Child()
obj1.speak()
```

```python
obj2=Parent()
obj2.speak()
```
