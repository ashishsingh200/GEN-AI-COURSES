```python
#example
class Avenger:
    def __init__(self,name,power,weapon,nationality='America'):
        self.name=name
        self.power=power
        self.weapon=weapon
        self.nationality=nationality
    def introduce(self): #here introduce is known as methods
        print(f'Hello I am {self.name}. I am from {self.nationality}')
    def carrying_weapon(self):
        print(f'I am carrying weapon {self.weapon}')

s1=Avenger('Spiderman','Spidy Sense', 'web-fluid')
s1.introduce()
s1.carrying_weapon()
```

```python
class Cartoon:
    def __init__(self,name,show,catchphrase):
        self.name=name
        self.show=show
        self.catchphrase=catchphrase
    def introduce(self):
        print(f'Hii my name is {self.name}')
    def speak(self):
        print(f'{self.name} saying {self.catchphrase}')
    def shows(self):
        print(f'{self.name} comes in {self.show}')
```

# Type of Methods

- Instance Method
    - it uses self to access and modify attributes

```python
class Avenger:
    def __init__(self,name,power):
        self.name=name
        self.power=power
    
    def show_details(self):
        print(f"I am {self.name} and I am having power {self.power}.")
```

```python
thor=Avenger('Thor','Thunder')
thor.show_details()
```

- Class Method
    - uses cls instead of self
    - abbreviations used *@classmethod*

```python
class Avengers:
    origin="Avenger"
    total_heroes=0

    def __init__(self,name,power):
        self.name=name
        self.power=power
        Avengers.total_heroes +=1
    
    @classmethod
    def origin_info(cls):
        print(f'My origin is {cls.origin}')
    @classmethod
    def count(cls):
        print(f'Total no. of avenger is {cls.total_heroes}')
    
    def intro(self):
        print(f'My name is {self.name} and my power is {self.power}')
        
```

```python
av1=Avengers('Iron Man','Genius Intellect')
av1.intro()
av1.origin_info()
```

```python
av2=Avengers('Thor','God of Thunder')
av2.intro()
av2.origin_info()
av2.count()
```

- Static Method
    - will niether use cls nor self
    - abbreviation used *@staticmethod*

```python
class Avengerss:

    @staticmethod
    def mission():
        print("All Avengers must Protect Earth at all cost...")
    
    origin='Avengers'

    def __init__(self,name,power):
        self.name=name
        self.power=power
    
    @classmethod
    def origin_info(cls):
        print(f"From {cls.origin}")

    def about(self):
        print(f'Name {self.name} \n Power: {self.power}')
    
        
```

```python
av1=Avengerss('Hulk','Gamma Radiation, Super Strength, High regeneration,Genius')
av2=Avengerss('Thor','Regenerative Power, God of thunder, Brute Strength')
av3=Avengerss('Iron Man','High Intelectual, Armor, Tech, Billionare')
av4=Avengerss('Spiderman','web, high senses, brute strength')
```

```python
heroes=[]
heroes.append(av1)
heroes.append(av2)
heroes.append(av3)
heroes.append(av4)

```

```python
for hero in heroes:
    hero.about()
    hero.origin_info()
print()
hero.mission()
```
