## In Python we get to see 3 types of Inheritance

    - Single Inheritance

    - Multilevel Inheritance
    
    - Multiple Inheritance

#### Single Inheritance --> The features or the Attributes from one parent class will be getting inherited into one child class.

### parent class (Properties --> atrributes and functions)
###             |
###             |
###         Child class

```python
class Avenger: # Parent Class
    def __init__(self,name,power):
        self.name=name
        self.power=power
    def introduction(self):
        print(f'Hello my name is {self.name}')

class Ironman(Avenger):
    ''' 
        Child Class to make a child class we have to make 
        one small parenthesis and inside parenthesis we have to
        write the name of parent class
    '''
    def __init__(self, name, power,weapon):
        super().__init__(name, power)
        self.weapon = weapon

    def weapon_details(self):
        print(f'I use {self.weapon}')
```

```python
iron=Ironman('Tony Stark', 'Genius Intellect','Nano Armor')
iron.introduction()
iron.weapon_details()
```

#### Multilevel Inheritance --> The properties or methods and attributes of the grand parent class and then the properties of the parent class will be inherited by the child class.

#### Grand Parent Class
#### |
#### Parent Class
#### |
#### Child Class


```python
class GrandParent:
    def __init__(self,nationality,baldness,anger_issues):
        self.nationality=nationality
        self.baldness=baldness
        self.anger_issues=anger_issues
    
    def introduce(self):
        print(f'Hello my Nationality is {self.nationality} and about my anger issues if u want to ask about then the answer is {self.anger_issues}')

class Parent(GrandParent):
    def __init__(self, nationality, baldness, anger_issues,hair,height,intelligence):
        super().__init__(nationality, baldness, anger_issues)
        self.hair=hair
        self.heights = height
        self.intelligence=intelligence

    def about_hair(self):
        print(f'I have {self.hair}.')
    def height(self):
        print(f'I am {self.heights} than my father.')

class Child(Parent):
    def __init__(self, nationality, baldness, anger_issues, hair, height, intelligence,charm):
        super().__init__(nationality, baldness, anger_issues, hair, height, intelligence)
        self.charm=charm
    def charm_details(self):
        print(f'About my charm if u guys want to know that wether if i am charming or not then the answer is {self.charm}')
```

```python
parent_1=Parent('Indian','No','Yes','Curly','taller','somewhat')
```

```python
parent_1.introduce()
```

```python
boy1=Child('Indian','No','High Anger issues','Curly hair','taller','Intelligent','Yes very much')
```

```python
boy1.introduce()
boy1.about_hair()
boy1.charm_details()
```

#### Multiple Inheritance --> In Multiple Inheritance the properties of the two or more parent class will be getting inherited into the child class

        Parent 1(Properties)   + parent2(Properties) + ...
                                |
                                |
                            Child Class

```python
class Ironman:
    def __init__(self,name,power,suit,weapon,nationality='American'):
        self.name=name
        self.power=power
        self.suit=suit
        self.weapon=weapon
        self.nationality=nationality
    def introduce(self):
        print(f'I am {self.name} and powers are {self.power}.I am {self.nationality}')

class CaptionAmerica:
    def __init__(self,name,power,suit,weapon,nationality='American'):
        self.name=name
        self.power=power
        self.suit=suit
        self.weapon=weapon
        self.nationality=nationality
    def weapon_details(self):
        print(f'My main weapon is {self.weapon}')
```

```python
class Hulk(Ironman,CaptionAmerica):
    def __init__(self, name, power, suit, weapon, anger_issues, nationality='American'):
        super().__init__(name, power, suit, weapon, nationality)
        self.anger_issues=anger_issues

    def anger_description(self):
        print(f"That's my secret, Captain: I am {self.anger_issues}.")

        
```

```python
hulku=Hulk('Prof. Banner','Brute Strength, Regeneratiion','Mark 85','Repulsor Beam, Sheild','always angry')
```

```python
hulku.introduce()
hulku.weapon_details()
hulku.anger_description()
```
