## 1.Create a dictionary price_dict where keys are product names and values are prices(integers or floats).Include atleast Six entries

```python
price_dict={'pen': 20,'note-book':40,'ear_buds':1500,'Bottle':90.56,'Shampoo':399,'highlighter':75.84}

print(price_dict)
```

## 2. Write a Small code block to :
            -- Update the price of an existing product.
            -- Remove a product by name (handle the case when the product does not exist).

            

```python
# o1 --updating the price of an existing product

price_dict['pen']=45
print(price_dict)

# 02 --removing the product by name
name='note-book'
if name in price_dict:
    price_dict.pop(name)
else: print('Enter the Correct Product')

print(price_dict)
```

## 3.Print the average price of all products(use only dictionary operations and basic arithmetics).

```python
price=price_dict.values()
print(f'All the prices are: {price}')
length=len(price_dict)
total=sum(price_dict.values())
average=total/length
print(f'Average Price of all products : {average:.2f}')
```

## Extra: Print the product with both the maximum and minimum prices.

```python
max_price=max(price_dict.values())
min_price=min(price_dict.values())
for name in price_dict:
    if price_dict[name]==max_price:
        max_product=name
    if price_dict[name]==min_price:
        min_product=name

print(f'Product name {max_product} has max price of {max_price}')
print(f'Product name {min_product} has max price of {min_price}')
```
