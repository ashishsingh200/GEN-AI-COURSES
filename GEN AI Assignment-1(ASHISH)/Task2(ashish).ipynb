```python
#adding new Categories in the set
#01-Duplicate Categories like Study Material
print(f'Before adding categories set items and length are: {categories_set} and {len(categories_set)}')
categories_set.add('Study Material') #adding duplicates items
print(f'After adding Duplicate item in categories set items and length are: {categories_set} and {len(categories_set)}')

# 02-Not a Duplicate  Categories.
print(f'Before adding categories set items and length are: {categories_set} and {len(categories_set)}')
categories_set.add('Snacks') #adding new item in the categories set
print(f'After adding categories set items and length are: {categories_set} and {len(categories_set)}')
```
