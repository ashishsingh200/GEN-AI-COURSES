## 1. using the product list and price_dict,create a list of tuples named catalog where each tuple is (product_name ,price, categoy)

```python
products=['pen','paper','copy','highlighter','bottle','charger','macbook']
price_dict={'pen':20,'paper':20.25,'copy':175.66,'highlighter':230.45,'bottle':80,'charger':2500,'macbook':71648}
category=['Stationary','Stationary','Stationary','Stationary','Utility','Electronics','Electronics']


catalog=[]
for product,cat in zip(products,category):
    catalog.append((product,price_dict[product],cat))
print(catalog)
```

## 2. From catalog, create a new dictionary category_to_products that maps each category to a list of product_names in that category

```python
category_to_products={}

for prod,price,cat in catalog:
    if cat in category_to_products:
        category_to_products[cat].append(prod)
    else:
        category_to_products[cat]=[prod]
print(category_to_products)
```

## 3.Print. all products that belong to the category that has the maximum number of products

```python
maximum=0
category=''
for cat in category_to_products:
    if len(category_to_products[cat])>maximum:
        maximum=len(category_to_products[cat])
        category=cat
print(category_to_products[category])
```
