```python
def apply_discount(price,discount_percent=5): #applying default discount percent at 5
    if discount_percent>60: #Assuring that discount_percent never exceeds 60%
        return print('Invalid Discount Rate!')
    else:
        discount=price * (discount_percent/100) #Calculating discount based on discount percent given
        dis_price=price-discount #calculating the final price after calculating discount
        print(dis_price) #printing the final result
```

```python
apply_discount(1000,10)
```

```python
apply_discount(500)
```

```python
apply_discount(3000,90)
```

```python
def factorial(n): #defining a function to find its factorial
    if n<0: # Prints an error when negative number is inserted
        print(f'{n} is a negative number.To find the factorial please enter positive number.')
    elif n==0 or n==1: return 1 #handles the edge cases
    else:
        return n*factorial(n-1) #this is the recursive function to find factorial of n
```

```python
factorial(5)
```

```python
factorial(0)
```

```python
factorial(-3)
```

```python
def process_prices(prices): # defining the function to get discounted and filtered price
    discounted_price=list(map(lambda x: x+(x*0.1),prices)) # using msp() + lsmbda to apply discount to all prices.
    filtered_prices=list(filter(lambda x: x>300,discounted_price)) # using filter to remove prices which is below or equal to 300
    return discounted_price, filtered_prices # returning the both list(discounted price and filters price)
 
discounted_price,filtered_price =process_prices([100,500,900,50,750])
print(f'The Discount price we get {discounted_price}')
print(f'After filtering dicounted price above 300 is {filtered_price}')
```
