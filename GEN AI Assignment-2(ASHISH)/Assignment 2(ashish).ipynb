```python
order_amount=input('Enther the order amount: ')#01
print('Your order amount is:', order_amount)
if not order_amount.isdigit():
    print("Enter the Valid amount...")
else:
    order_amount=int(order_amount)
    
    if order_amount<0 :#03
        print('Order Amount can not be negative..')
        exit()
    elif order_amount>=2000:#02
        print("You are eligible for '15%' discount.")
        discount=order_amount*.15
        final_amount=order_amount-discount
    elif order_amount<2000 and order_amount>=1500:
        print("You are eligible for '10%' discount.")
        discount=order_amount*.10
        final_amount=order_amount-discount
    elif order_amount<1500 and order_amount>=1000:
        print("You are eligible for '7%' discount.")
        discount=order_amount*.07
        final_amount=order_amount-discount
    else:
        print("You are eligible for '0%' discount.")
        final_amount=order_amount
    #Extra
    tax=final_amount*.05
    total=final_amount+tax

print('The Subtotal amount is:',final_amount)
print('The tax(5%) in your order is:',tax)
print('The Final amount is:',total)
```

```python
orders=[1200,2500,800,1750,3000]
total_revenue=0
count=0
#01
for order_amount in orders:
    print('Your order amount is:', order_amount)
    if order_amount<0 or type(order_amount)==str:#03
        print("Enter the Valid amount...")
    elif order_amount>=2000:#02
        print("You are eligible for '15%' discount.")
        disc=0.15
        discount=order_amount* disc
        final_amount=order_amount-discount
    elif order_amount<2000 and order_amount>=1500:
        print("You are eligible for '10%' discount.")
        disc=0.10
        discount=order_amount* disc
        final_amount=order_amount-discount
    elif order_amount<1500 and order_amount>=1000:
        print("You are eligible for '7%' discount.")
        disc=0.07
        discount=order_amount* disc
        final_amount=order_amount-discount
    else:
        disc=0
        print("You are eligible for '0%' discount.")
        final_amount=order_amount
    if disc>0: count +=1
    total_revenue +=final_amount
    print(f'order_amount: {order_amount:.2f} --> discount(%): {disc} --> final_amount: {final_amount}')
print(f'Total Revenue : {total_revenue}')

print(f'The Number of products that recieved a discount greater than 0: {count}')
```

```python
orders=[]
count=0
while True:
    print("----Menu----")
    print('1 -- Add order amount')
    print('2 -- show all orders')
    print(' q to quit')

    choice=input('Enter the Choice you want!')

    if choice=='q':
        break
    elif choice !='1' and choice !='2':
        print("Enter the correct choice ether 1 or 2 is available for now...")
        continue
    elif choice=='1':
        order_amount=int(input('Enter the order amount: '))
        if order_amount<0:
            print('Please enter the correct amount.')
            continue
        orders.append(order_amount)
        print(f'Order of {order_amount} added successfully')
    elif choice=='2':
        if len(orders)==0:
            print('No Orders yet.')
            continue
        total_revenue=0

        for order_amount in orders:
            print('Your order amount is:', order_amount)
            if order_amount<0 or type(order_amount)==str:#03
                print("Enter the Valid amount...")
            elif order_amount>=2000:#02
                print("You are eligible for '15%' discount.")
                disc=0.15
                discount=order_amount* disc
                final_amount=order_amount-discount
            elif order_amount<2000 and order_amount>=1500:
                print("You are eligible for '10%' discount.")
                disc=0.10
                discount=order_amount* disc
                final_amount=order_amount-discount
            elif order_amount<1500 and order_amount>=1000:
                print("You are eligible for '7%' discount.")
                disc=0.07
                discount=order_amount* disc
                final_amount=order_amount-discount
            else:
                disc=0
                print("You are eligible for '0%' discount.")
                final_amount=order_amount
            if disc>0: count +=1
            total_revenue +=final_amount
            print(f'order_amount: {order_amount:.2f} --> discount(%): {disc} --> final_amount: {final_amount}')
        print(f'Total Revenue : {total_revenue}')
            
```
