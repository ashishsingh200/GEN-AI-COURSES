```python
# Step 01 --> creating the list of sales amounts
sales=[1200,450,980,1500,3000]

# Step 02 --> using 'w' to write the data inside the text file. if that text file is not present then it will create one.
with open('sales_data.txt','w') as file:
    for sale in sales:
        file.write(str(sale) + '\n')

# Step 03 --> printing the context from the text file using 'r' mode
with open('sales_data.txt','r') as file:
    print(file.read())

# Step 04 --> Extra: writing the data in comma seperated format instead of seprate lines
with open('sales_data.txt','w') as file:
    file.write(','.join(str(sale) for sale in sales))

# Notes: since we are using with open there is no need to close the file as it automatically closes it
```

```python
# Step 01 --> Storing the new sales data into a list new_sales
new_sales=[5000,2500,1700]

# Step 02 --> Appending those new sale to the same file 
with open('sales_data.txt','a') as file:
    for sale in new_sales:
        file.write(str(sale) + '\n')

# Step 03 --> After appending opening the file and printing the entire updated file
with open('sales_data.txt','r') as file:
    print(file.read())
# Step 04 --> Extra: Printing the total number of lines using len()
with open('sales_data.txt','r') as file:
    lines=file.readlines()
    print(len(lines))
```

```python
# Step 01 --> Creating Variables
total_sale=0
average_sale=0
highest_sale=0
lowest_sale=999999

# Step 02 --> Reading all sales value from sales_data.txt using 'r' mode
with open('sales_data.txt','r') as file:
    line=file.readlines()
    for sale in line:
        total_sale += int(sale) # finding the total sale
        highest_sale=max(highest_sale,int(sale)) #finding highest sale
        lowest_sale=min(lowest_sale,int(sale)) #finding lowest sale
    average_sale=total_sale/len(line) #finding average sale

print(f'Total Sale: {total_sale}')
print(f'Highest Sale: {highest_sale}')
print(f'Lowest Sale: {lowest_sale}')
print(f'Average Sale: {average_sale}')
        
    
```

```python
# Step 01 --> Creating variables
avg=0
discounted_price=0
total_discount=0

# Step 02 --> Creating dictionary name prices and storing the variables
prices={
    "Mouse":500,
    "Keyboard": 800,
    "Monitor":7000,
    "Pendrive": 400,
    "Camera":5000
}
# Step 03 --> Asking the user to provide the discount percentage
discount_percent=int(input('Enter the discount percentage: '))

# Step 04 --> Writing the contents as per the questions in file discount_price.txt
with open('discount_price.txt','w') as file: 
    file.write('Product     | Price     | Discounted Price \n')
    for product,price in prices.items():
        discounted_price=price + ((price * discount_percent)/100)
        total_discount += discounted_price # finding total discount

        # Step 05 --> writing the details inside the file
        file.write(f'{product}   {price}     {discounted_price}\n')
    file.write('\n')
    # Step 06 --> Writing the summary
    file.write(f'Total Items: {len(prices)}\n')
    file.write(f'Avereage Discounted Price: {total_discount/len(prices)}')

with open('discount_price.txt', 'r') as file:
    print(file.read())
```
