```python
class Product: # defining a class Product having attributes name price categofy
    def __init__(self,name,price,category,discount=0): # for extra question making discount atttributes as well and defaulting it to 0
        self.name=name
        self.price=price
        self.category=category
        self.discount=discount
    
    def get_info(self): #creating get_info method  to print the product details
        print(f'The name of the product is {self.name} its price is {self.price} comes under category {self.category}.')

    def apply_discount(self): # creating another method called apply discount to print the discount amount as stated in extra question
        if self.discount !=0:
            sum=0
            sum= self.price - (self.price*self.discount)/100
            print(f'After getting a discount of {self.discount} you need to pay {sum}')
        else:
            print(f'No Discount you have to pay full price {self.price}')
```

```python
# creating 2 objects namely obj1 and obj2
obj1=Product('pen',400,'Stationary')
obj2=Product('Mac',72800,'Elevtronics',15)
```

```python
# as stated in the question applying get_info method and also to solve the extra question, calling apply_discount method
obj1.get_info()
obj1.apply_discount()
```

```python
obj2.get_info()
obj2.apply_discount()
```

```python
class ElectronicProduct(Product):
    def __init__(self, name, price, category,waranty_years, discount=0):
        super().__init__(name, price, category, discount)
        self.waranty_years=waranty_years
    
    def get_info(self):
        super().get_info()
        print(f'It has a discount of {self.discount}% and a waranty period of {self.waranty_years} years.')
```

```python
obj3=ElectronicProduct(
    name="Samsung Galaxy", 
    price=45000, 
    category="Smartphone", 
    waranty_years=2
)
obj3.get_info()
obj3.apply_discount()
```
